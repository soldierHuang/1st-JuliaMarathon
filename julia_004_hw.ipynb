## Day 004 作業

今天的內容提到整數及浮點數型別之最小值和最大值

### 作業：範例程式示範了 Int64 及 Float64 的最小值和最大值，作業請列出並觀察其他整數及浮點數型別的最小值和最大值。

呼叫 `typemin()` 及 `typemax()` 函式得到各型別的最小值和最大值。

列出 Int128, Int32, Int16, Int8, UInt128, UInt64, UInt32, UInt16, UInt8 整數型別的最小值及最大值

|是否有正負號|型別|<div style="width:120px">最大值</div>|<div style="width:120px">最小值</div>|
|---|---|---|---|
|有|Int128, Int64, Int32, Int16, Int8|$2^{位元數-1}-1$|$-2^{位元數-1}$|
|無|UInt128, UInt64, UInt32, UInt16, UInt8|$2^{位元數}-1$|$0$|

```julia
!julia -e '\
println(" Int128: 最小值 $(typemin(Int128)), 最大值 $(typemax(Int128))");\
println(" Int32: 最小值 $(typemin(Int32)), 最大值 $(typemax(Int32))"); \
println(" Int16: 最小值 $(typemin(Int16)), 最大值 $(typemax(Int16))"); \
println(" Int8: 最小值 $(typemin(Int8)), 最大值 $(typemax(Int8))"); \
println(" UInt128: 最小值 $(typemin(UInt128)), 最大值 $(typemax(UInt128))"); \
println(" UInt64: 最小值 $(typemin(UInt64)), 最大值 $(typemax(UInt64))"); \
println(" UInt32: 最小值 $(typemin(UInt32)), 最大值 $(typemax(UInt32))"); \
println(" UInt16: 最小值 $(typemin(UInt16)), 最大值 $(typemax(UInt16))"); \
println(" UInt8: 最小值 $(typemin(UInt8)), 最大值 $(typemax(UInt8))"); \'
```

## Float 

|型別|最小值|最大值|
|---|---|---|
|`Float64`|-Inf|Inf|
|`Float32`|-Inf 或 -Inf32|Inf 或 Inf32|
|`Float16`|-Inf 或 -Inf16|Inf 或 Inf16|
